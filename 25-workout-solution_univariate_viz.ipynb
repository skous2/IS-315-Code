Anneli, a teacher, wants to know if there is a relationship between reading percent and how students do on assessments. She downloaded a CSV of the studentsâ€™ reading percent and average assessment score. Before completing any statistical analysis, Anneli wants to use histograms to see if there is a relationship. Make a double histogram with the data to see if there is any relationship. 

To make the graph easier to read, include a key that shows what each line represents.

```python
###using matplotlib

plt.hist(grades['Average assesment performance'], bins=int(round(grades['Reading completion'].count()**(1/3), 0)), label='assessment')
plt.hist(grades['Reading completion'], bins=int(round(grades['Reading completion'].count()**(1/3), 0)), label = 'reading completion')
plt.ylabel('count of students')
plt.xlabel('percentage')
plt.legend(loc="upper left")
```

```python
### Using seaborn
import seaborn as sns

ax = sns.histplot(data=grades['Average assesment performance'], kde=True, color="orange")
ax = sns.histplot(grades['Reading completion'], kde=True)

ax.set(yticks=[])
ax.set(xticks=[])
ax.set_xlabel('')
ax.legend(labels=['Average assessment', 'Reading completion'])
sns.despine(ax=ax)
```
