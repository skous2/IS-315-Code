Search online to find out how to change the default correlation algorithm from Pearson to Spearman and then Kendall in the Scipy stats package. 

Create a DataFrame with the results of the Pearson, Spearman and the Kendall correlations and p-values for the heart attack dataset

It should look like this:

| . | Pearson r | Pearson p | Spearman r | Spearman p | Kendall tau | Kendall p |
| - | - | - | - | - | - | - | 
|cp | .503 | .000 | .517 | .000 | .486 | .000 |
| etc | # | #| # | # | # | # | # |  

```python
import pandas as pd
from scipy import stats

df = pd.read_csv('https://www.ishelp.info/data/heart_attack_clean.csv')
```

```python
columns =[]

pearsonr = []
pearsonp = []

spearmanr = []
spearmanp = []

kendalltau = []
kendallp = []

for col in df.columns:
    columns.append(col)
    if col != 'num:':
        pearson = stats.pearsonr(df.num, df[col])
        spearman = stats.spearmanr(df.num, df[col])
        kendall = stats.kendalltau(df.num, df[col])
        
        pearsonr.append(pearson[0])
        pearsonp.append(pearson[1])

        spearmanr.append(spearman[0])
        spearmanp.append(spearman[1])

        kendalltau.append(kendall[0])
        kendallp.append(kendall[1])


statsdf = pd.DataFrame(columns=['Pearson r', 'Pearson p', 'Spearman r', 'Spearman p', 'Kendall tau', 'Kendall p'])

print(statsdf)

statsdf['Pearson r'] = pearsonr
statsdf['Pearson p'] = pearsonp
statsdf['Spearman r'] = spearmanr
statsdf['Spearman p'] = spearmanp
statsdf['Kendall tau'] = kendalltau
statsdf['Kendall p'] = kendallp

statsdf.index = columns

statsdf.round(4)

```
